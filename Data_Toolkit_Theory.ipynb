### **6. What are the differences between NumPy arrays and Python lists?**


* *NumPy Array:* A fixed-size, homogeneous data structure provided by the NumPy library, optimized for numerical computations.
* *Python List:* A flexible, built-in data structure in Python that can hold heterogeneous data types.

#### **Differences Between NumPy Arrays and Python Lists**

| **Feature**          | **NumPy Arrays**                                        | **Python Lists**                                 |
|-----------------------|---------------------------------------------------------|-------------------------------------------------|
| **Data Type**         | Homogeneous: All elements must be of the same data type.| Heterogeneous: Can contain elements of different data types. |
| **Performance**       | Faster due to low-level optimizations (C implementation).| Slower due to dynamic typing and Python's interpreter overhead. |
| **Memory Efficiency** | More memory-efficient due to compact storage.           | Consumes more memory as it stores additional metadata. |
| **Operations**        | Supports element-wise operations directly (e.g., addition, multiplication). | Requires explicit loops or list comprehensions for similar operations. |
| **Dimensionality**    | Supports multi-dimensional arrays (e.g., matrices, tensors). | Native lists are one-dimensional (can nest lists for higher dimensions, but it's inefficient). |
| **Mathematical Functions** | Comes with built-in mathematical and statistical functions (e.g., `np.mean`, `np.sum`). | Requires manual implementation or external libraries (e.g., `math` module). |

---

### **Why is NumPy Better for Numerical Computations?**

1. **Speed:**  
   NumPy is faster because it uses low-level C implementations. Operations like addition or multiplication are performed in parallel at the hardware level.  

2. **Efficiency:**  
   NumPy stores data in contiguous memory blocks, reducing memory overhead compared to Python lists.  

3. **Versatility:**  
   NumPy arrays support advanced mathematical operations like matrix multiplication, linear algebra, and broadcasting, which are not natively supported by Python lists.  

---



```python
import numpy as np

# Python list allows mixed types
list_example = [1, "two", 3.0]
print("Python List:", list_example)

# NumPy array requires homogeneous data types
numpy_array = np.array([1, 2, 3])
print("NumPy Array:", numpy_array)
```

```python
#EFFICIENCY
import time

# Large list and NumPy array
large_list = list(range(1_000_000))
large_array = np.array(large_list)  #converting the same list in array

# Python list operation
start_time = time.time()
list_result = [x * 2 for x in large_list]
time_taken_l = time.time() - start_time
print(f"Python List Time: {(time_taken_l):.5f} seconds")

# NumPy array operation
start_time = time.time()
array_result = large_array * 2
time_taken_a = time.time() - start_time
print(f"NumPy Array Time: {(time_taken_a):.5f} seconds")

compared_time = round((time_taken_l / time_taken_a))
print(f"\nNumpy arrays are approximately {compared_time} times faster.")
```

### **9. How does Matplotlib differ from Plotly?**


* **Matplotlib:** A static plotting library for creating 2D graphs (Though it supports animations but the code becomes complicated).
* **Plotly:** A library for interactive, web-based, 2D, and 3D visualizations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(0, 10, 50)  # 50 points evenly spaced between 0 and 10
y = 3 * x + 5               # Line equation y = 3x + 5

# Create scatter plot and line plot
plt.scatter(x, y, label="Data Points")  # Scatter plot
plt.plot(x, y, color="red", label="Line: y = 3x + 5")  # Line plot

# Add labels, title, and legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter and Line Plot")
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Generate data
x = np.linspace(0, 10, 50)  # 50 points evenly spaced between 0 and 10
y = 3 * x + 5               # Line equation y = 3x + 5

# Create the figure
fig = go.Figure()

# Add scatter plot
fig.add_trace(go.Scatter(x=x, y=y, mode='markers', name='Data Points'))

# Add line plot
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Line: y = 3x + 5', line=dict(color='red')))

# Customize layout
fig.update_layout(
    title="Scatter and Line Plot with Plotly",
    xaxis_title="X-axis",
    yaxis_title="Y-axis",
    legend_title="Legend"
)

# Show the plot
fig.show()
```

### 17. Explain the difference between apply() and map() in Pandas


#### Difference between `apply()` and `map()` in Pandas

Both `apply()` and `map()` are used to apply functions to data in Pandas, but they are used in different contexts and have different behaviors.

1. **Usage Context**

- **`apply()`**:
  - Can be used with both DataFrames or Series.
  - It can apply a function along either axis (rows or columns) of a DataFrame or to an entire Series.
  - Can be used for more complex operations, including applying a function to multiple columns or performing aggregation.

- **`map()`**:
  - Primarily being used with Series objects.
  - It applies a function element-wise to the values in a Series.
  - Typically used for substituting values or transforming data based on a mapping (like replacing values with a dictionary).

2. **Function Application**

- **`apply()`**:
  - Can apply any function that operates on an entire row or column.
  - Supports both row-wise and column-wise operations in DataFrames using the `axis` argument.
  

- **`map()`**:
  - Works element-wise and is best suited for simple transformations (e.g., replacing or modifying individual elements).
  - Can be used with dictionaries, functions, or Series for value replacements.

3. **Performance**

- **`apply()`**:
  - Generally slower for element-wise operations compared to `map()` since it is more flexible and can handle more complex functions.
  
- **`map()`**:
  - More efficient for element-wise operations, especially when using a mapping (dictionary or Series), since it is optimized for simple transformations.

4. **Handling Missing Data**

- **`apply()`**:
  - Can handle missing data (`NaN`) in more complex ways, depending on the function being applied.
  
- **`map()`**:
  - Handles missing data by returning `NaN` for unmatched values when used with a dictionary or Series for mapping.


**To Conclude**:
- We use `apply()` when we need to perform operations that involve rows, columns, or more complex transformations in a DataFrame or Series.
- And we use `map()` when we need to apply a function element-wise on a Series or replace values based on a dictionary, Series, or function.


```python
#EXAMPLE OF APPLY()
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['Sum'] = df.apply(lambda row: row['A'] + row['B'], axis=1)
df
```

```python
#EXAMPLE OF MAP()
series = pd.Series([1, 2, 3])
result = series.map(lambda x: x * 2)
result
```

### 19. How does Pandas simplify time series analysis


Analyzing and forecasting data indexed or depens on time, such as stock prices or weather trends are known as as Time series analysis.

Pandas provides powerful tools for working with time series data, simplifying the process of analyzing, manipulating, and visualizing time-based information.

**Key Features of Pandas for Time Series Analysis:-**
1. Datetime Indexing: Automatically parses and sets datetime objects as index.

2. Resampling: Aggregates data over time intervals (e.g., weekly, monthly).

3. Shifting: Allows time-based transformations by shifting data forward or backward in time useful for calculating differences.
4. Time Zone Handling: Supports timezone-aware data manipulation.
5. Date Range Generation: Easily create date ranges for indexing or generating time periods.


```python
#GENERATING A TIME SERIES DATA

import pandas as pd
import numpy as np

# Generate a date range from January 1, 2025 to January 10, 2025, with daily frequency
dates = pd.date_range('2025-01-01', periods=10, freq='D')

# Creating a random tdata points
data = np.random.randn(10) * 10 + 50

# Create a DataFrame with the date index
df = pd.DataFrame(data, index=dates, columns=['Value']) #time serues data

df
```

```python
# aggregating the data to weekly frequency and take the mean
weekly_data = df.resample('W').mean()       #calculating the weekly mean

weekly_data
```

```python
# shifting data


df['Shifted'] = df['Value'].shift(1)        #1 shifts the values by 1 day forward

df
# now to calculate the difference in daily time interval we can
df['Difference'] = df['Value'] - df['Shifted']

df
```

### 20. What is the role of a pivot table in Pandas


A pivot table summarizes and organizes data, enabling multi-dimensional analysis by aggregating and rearranging rows and columns.


Pivot tables are used to quickly summarize large datasets.
Identify patterns or trends in the data.
Especially when working with categorical and numerical data.

```python
import pandas as pd

# Sample data for 1 week sales of some products
data = {
    'Date': pd.date_range('2025-01-01', periods=6, freq='D'),
    'Region': ['East', 'West', 'East', 'West', 'East', 'West'],
    'Product': ['A', 'A', 'B', 'B', 'A', 'B'],
    'Sales': [100, 150, 200, 180, 120, 160]
}

# Converting to DataFrame
df = pd.DataFrame(data)

df
```

```python
#creating a pivot table that will show the total sales for each region and product for the week

# Creating the pivot table
pivot_table = df.pivot_table(values='Sales', index='Region', columns='Product', aggfunc='sum')

pivot_table
```

```python
# we can also use a list of functions in aggfunc argument to generate more summarization at once

pivot_table_multiple = df.pivot_table(values='Sales', index='Region', columns='Product', aggfunc=['sum', 'mean'])
pivot_table_multiple
```

### 21. Why is NumPy’s array slicing faster than Python’s list slicing
